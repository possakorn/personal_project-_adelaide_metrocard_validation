## Import dataset

```python
import urllib
import urllib.request
import json
import pandas as pd
import numpy as np
```

```python
resources_list = {
	'2024_q2': 'fb924e9e-9ea4-4d30-8f0c-51e556f0bb36',
	'2024_q1': '8986f2ef-aeca-4645-9443-1b9b5db07c1a'
}



def download_df(resource_id):
	url = f'https://data.sa.gov.au/data/api/3/action/datastore_search?resource_id={resource_id}&limit=1000000'
	# Step 1: Request API
	with urllib.request.urlopen(url) as url:
		s = url.read()
	# Step 2: Decode the byte string
	json_string = s.decode('utf-8')
	data = json.loads(json_string)
	# Step 3: Change to pandas dataframe
	df = pd.DataFrame(data['result']['records'])
	return df




df = {}
for name, rs_id in resources_list.items():
    df[name] = download_df(rs_id)
```

```python
BASE_URL = f'https://data.sa.gov.au/data'
INITIAL_URL = "/api/3/action/datastore_search?resource_id"
# INITIAL_URL = "/api/3/action/datastore_search?resource_id=d54f7465-74b8-4fff-8653-37e724d0ebbb"
LIMIT = 1000000
import requests
def get_all(resource_id) -> list:
    result = []
    resp = requests.get(f"{BASE_URL}{INITIAL_URL}={resource_id}&limit={LIMIT}")
    js = resp.json()["result"]
    result.extend(js["records"])
    while "_links" in js and "next" in js["_links"]:
        resp = requests.get(BASE_URL + js["_links"]["next"])
        js = resp.json()["result"]
        result.extend(js["records"])
        print(js["_links"]["next"]) # just so you know it's actually doing stuff
        if len(js["records"]) < LIMIT:
            # if it returned less records than the limit, the end has been reached
            break
    return result

data = get_all('8986f2ef-aeca-4645-9443-1b9b5db07c1a')
data
```
